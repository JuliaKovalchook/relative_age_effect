```python
'''
Function for open file
'''

def open_pages(file_name):
    file_exists = exists(file_name)
    if file_exists==True:
        done_list=pd.read_csv(file_name, index_col=0)   #open the file with passed links 
        #Don't add Category pages. It is better to go through the extra pages of the category than not to get all the links from it
        print('File %s exists and have %s elements in total' % (file_name,len(done_list)) )
    else:
        done_list=pd.DataFrame(columns=['Link'])
        print('File %s not exists' % (file_name))
    return done_list
```

### Part 2

Convert the date to the desired format

```python
'''
Open a set of links
'''
        
# open file with links and person's birth
df_person_info=open_pages('Perosn.csv')
df_person_info
```

```python
'''
Function to get the month number
'''


def change_month(x):
    if x=='січня':
        month_number=1
    elif x=='лютого':
        month_number=2
    elif x=='березня':
        month_number=3
    elif x=='квітня':
        month_number=4
    elif x=='травня':
        month_number=5
    elif x=='червня':
        month_number=6
    elif x=='липня':
        month_number=7
    elif x=='серпня':
        month_number=8
    elif x=='вересня':
        month_number=9
    elif x=='жовтня':
        month_number=10
    elif x=='листопада':
        month_number=11
    elif x=='грудня':
        month_number=12
    
    else: month_number=0
    return month_number

def period(x):
    if x in (9,10,11):
        period='4 autumn'
    elif x in (12,1,2): 
        period='1 winter'   
    elif x in (3,4,5): 
        period='2 spring' 
    elif x in (6,7,8): 
        period='3 summer'  

    else: period='ERROR'
    return period

```

```python
df_person_info['Year'].unique()
```

```python
# Change the order of month and year if necessary

df_person_info[df_person_info['Month'].str.contains('\d{4}')]
df_person_info.loc[lambda x: x['Month'].str.contains('\d{4}'), 'Month'] = df_person_info['Year'] +' and ' + df_person_info['Month'] #concat Year and MOnth
df_person_info.loc[lambda x: x['Month'].str.contains('and'), 'Year'] = df_person_info['Month'].apply(lambda x: re.sub('.*and', '', x).strip()) #replace all befor #and#
df_person_info.loc[lambda x: x['Month'].str.contains('and'), 'Month'] = df_person_info['Month'].apply(lambda x: re.sub('\sand.*', '', x).strip()) #replace all after #and#


df_person_info
```

```python
df_person_info['Year'].unique()
```

```python
'''
Apply functions to get the month number
'''

#drop data about too young and too old people
df_person_info['Year']=df_person_info['Year'].astype(int)
df_person_info=df_person_info[df_person_info['Year']>1950]
df_person_info=df_person_info[df_person_info['Year']<2007]

#get period of month
df_person_info['short month']=df_person_info['Month'].apply(lambda x: re.sub('\d', '', x).strip())
df_person_info['month_number']=df_person_info['short month'].apply(lambda x: change_month(x))
df_person_info['period']=df_person_info['month_number'].apply(lambda x: period(x))

df_person_info
```
